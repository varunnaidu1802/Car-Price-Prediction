```python
dataset=df[['Year','Selling_Price','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']]
```

```python
dataset
```

```python
dataset
```

```python
dataset['Current_Year']=2022
```

```python
dataset['no.of_year']=dataset['Current_Year']-dataset['Year']
```

```python
dataset
```

```python
dataset.drop(['Year'],axis=1,inplace=True)
```

```python
dataset
```

```python
dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
dataset.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
dataset.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
dataset
```

```python
X= dataset.drop(['Selling_Price'],axis=1)
Y= dataset['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3 )
```

Linear Regression

```python
lreg_model= LinearRegression()
```

```python
lreg_model.fit(X_train,Y_train)
```

Model Evaluation

```python
training_dataprediction=lreg_model.predict(X_train)
```

Using Chi-Square Method

```python
import scipy.stats as stats
```

```python
chi_square_test_statistic, p_value = stats.chisquare(Y_train,training_dataprediction)
```

```python
print('chi_square_test_statistic is : ' +
      str(chi_square_test_statistic))
print('p_value : ' + str(p_value))
```

```python
plt.scatter(Y_train,training_dataprediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted_Price")
plt.title("Car Prediction")
plt.show()
```

```python
test_dataprediction=lreg_model.predict(X_test)
```

```python
chi_square_test_statistic, p_value = stats.chisquare(Y_test,test_dataprediction)
print('chi_square_test_statistic is : ' +
      str(chi_square_test_statistic))
print('p_value : ' + str(p_value))
```

```python
plt.scatter(Y_test,test_dataprediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted_Price")
plt.title("Car Prediction")
plt.show()
```
